```python
import pandas as pd
import dfgcompare.dataset.hospital as sampleData
from dfgcompare import dfg
from dfgcompare.dfg import stochastic as stoc_dfg
```

```python
df = sampleData.getData()
df_emergency = df[df['isEmergency']]
df_not_emergency = df[df['isEmergency']==False]
```

```python
dfg_full = dfg.computeDFG(df, 'caseid', 'activity', 'order')
dfg_emergency = dfg.computeDFG(df_emergency, 'caseid', 'activity', 'order')
dfg_not_emergency = dfg.computeDFG(df_not_emergency, 'caseid', 'activity', 'order')
```

```python
g_full = dfg.drawDFG(dfg_full)
g_emergency = dfg.drawDFG(dfg_emergency)
g_not_emergency = dfg.drawDFG(dfg_not_emergency)
```

```python
dfg_stoc_emergency = stoc_dfg.computeStochasticDFG(dfg_emergency)
dfg_stoc_not_emergency = stoc_dfg.computeStochasticDFG(dfg_not_emergency)
```

```python
res = stoc_dfg.compareStochasticDFGs(dfg_stoc_emergency, dfg_stoc_not_emergency, 0.2)
```
